```python
df_filter = df[
    (
        (df["Built Year"] <= 2028)
        | (df["Built Year"].isnull())
    ) 
    & (df["Kitchens"] <= 6) 
    & (df["Store Rooms"] <= 6) 
    & (df["Servant Quarters"] <= 6)
]

df_filter
```

```python
display(df.shape)
df_filter.shape
```

```python
price_split = df_filter["Price"].str.split(expand=True)
price_split[1].isnull().value_counts()
```

```python
missing_unit_mask = price_split[1].isnull()
df_filter = df_filter[~missing_unit_mask]
```

```python
price_num = pd.to_numeric(price_split[0])

price_mapping = {
    "Crore": 10_000_000,
    "Lakh": 100_000,
    "Arab": 1_000_000_000
}

price_multiplier = price_split[1].map(price_mapping)

price_pkr = price_num * price_multiplier

df_filter["Price_PKR"] = price_pkr
df_filter["Price_PKR"].isnull().sum()
```

```python
area_split = df_filter["Area"].str.split(expand=True)
display(area_split.isnull().sum())

area_num = pd.to_numeric(area_split[0])

area_mapping = {
    "Marla": 1,
    "Kanal": 20,
}

area_multiplier = area_split[1].map(area_mapping)

area_Marla = area_num * area_multiplier

df_filter["Area_Marla"] = area_Marla

df_filter[["Area_Marla", "Area"]]
```

```python
df_filter[["Price_PKR", "Area_Marla"]].describe()
```

```python
df_filter = df_filter[df_filter["Area_Marla"] != 0]
print(df_filter.shape)
```

```python
print(df_filter.shape)
print(df_filter.columns.tolist())
```

```python
bins = [1959, 2015, 2020, 2023, 2025, 2031]
labels = ['Pre-2015', '2015-2020', '2021-2023', '2024-2025', '2026+']
df_filter['Built_Year_Bucket'] = pd.cut(df_filter['Built Year'], bins=bins, labels=labels)
display(df_filter.groupby('Built_Year_Bucket')['Price_PKR'].agg(['count', 'mean', 'median']))
display(df_filter[df_filter['Built Year'].isnull()]['Price_PKR'].describe())
df_filter[df_filter['Built Year'].notnull()]['Price_PKR'].describe()
```

```python
df_filter.groupby('Built_Year_Bucket')['Area_Marla'].agg(['count', 'mean', 'median'])
```

```python
median_year = df_filter['Built Year'].median()

df_filter['Built_Year_Missing'] = df_filter['Built Year'].isnull().astype(int)
df_filter['Built Year'] = df_filter['Built Year'].fillna(median_year)
```

```python
display(df_filter['Built Year'].isnull().sum())
df_filter['Built_Year_Missing'].value_counts()
```

```python
df_filter
```

```python
df_filter['Bedrooms'] = pd.to_numeric(df_filter['Bedrooms'], errors='coerce')
df_filter['Bathrooms'] = pd.to_numeric(df_filter['Bathrooms'], errors='coerce')

df_filter['Bedrooms'] = df_filter['Bedrooms'].fillna(df_filter['Bedrooms'].median())
df_filter['Bathrooms'] = df_filter['Bathrooms'].fillna(df_filter['Bathrooms'].median())
```

```python
df_features = df_filter[["Location", "Area_Marla", "Bedrooms", "Bathrooms", "Built Year", 
                         "Kitchens", "Store Rooms", "Servant Quarters",]]
```

```python
df_filter = df_filter[df_filter["Built Year"] < 2026]
```

```python
first_part = df_features["Location"].str.split(",").str[0]
second_part = df_features["Location"].str.split(",").str[1].str.strip()
condition = first_part.str.contains(" - ")
```

```python
import numpy as np

df_features["Location_Clean"] = np.where(condition, second_part, first_part)
df_features[["Location", "Location_Clean"]].sample(10, random_state=42)
```

```python
df_features["Location_Clean"].nunique()
```

```python
df_features["Location_Clean"].value_counts().tail(20)
```

```python
location_counts = df_features["Location_Clean"].value_counts()

common_locations = location_counts[location_counts >= 20].index

df_features["Location_Clean"] = np.where(
    df_features["Location_Clean"].isin(common_locations),
    df_features["Location_Clean"],
    "Other"
)

print(df_features["Location_Clean"].nunique())
df_features["Location_Clean"].value_counts().head(20)
```

```python
df_model = pd.get_dummies(
    df_features,
    columns=["Location_Clean"],
    dtype=int
)

print(df_model.shape)
df_model.head()
```

```python
df_model.to_csv("../data/processed/cleaned_houses.csv", index=False)
```
